```python
#문제3
class Counter:
    def __init__(self,cnt,stp=1):
        self.count=cnt
        self.step=stp
    def __str__(self):
        return "[Count(step:"+str(self.step)+")]"+str(self.count)
    def incr(self):
        self.count+=self.step
    def __call__(self):
        return self.incr()
    def __add__(self,b):
        return Counter(self.count+b,self.step)
    def __sub__(self,b):
        return Counter(self.count-b,self.step)
    def __gt__(self,b):
        return self.count>b
    def __lt__(self,b):
        return self.count<b
    def __eq__(self,b):
        return self.count==b
    def __ne__(self,b):
        return self.count!=b
c=Counter(10)
d = Counter(10, 2)
print(c)
print(d)
c.incr()
d.incr()
print(c)
print(d)
c()
d()
print(c)
print(d)
c+=5
d-=5
print(c)
print(d)
print(c>10)
print(d>10)
print(c<10)
print(d<10)
print(c==17)
print(d!=9)
```

MySet이란 집합형태를 만드려는 의도로 만들어진 클래스입니다.   
eliminate_duplicate(l) 은 중복된 값들을 제거하기 위한 함수로 정적메소드입니다. 이 함수는 인스턴스 객체와 무관하게 클래스 이름공간에 존재하는 메소드로 메소드 이름을 이용하여 직접 호출할수 있습니다. 인자로 리스트 l을 받으면 리스트 s에 반복문을 이용해   리스트 l 에서 값하나씩 e로 빼서 중복되지않은 값들만 s에 추가한뒤 그 리스트를 다시 리스트 l에 넣습니다.  
__init__ 인 생성자는 리스트를 받아 그 값을 self에 저장한뒤 eliminate_duplicate(self)를 이용해 중복된 값을 제거합니다.  
__str__ 메소드를 이용해 class에 값인 리스트를 출력시 보기 좋게 출력되게 하였다.

```python
#문제5
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}" 
        return result
       
    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    def __or__(self,l):
        lista=[]
        for e in self:
            lista.append(e)
        for f in l:
            lista.append(f)    
        
        lista.sort()
        MySet.eliminate_duplicate(lista)
        return MySet(lista)
    def __and__(self,l):
        lista=[]
        for e in self:
            if e in l:
                lista.append(e)
        lista.sort()
        MySet.eliminate_duplicate(lista)        
        return MySet(lista)
    def __sub__(self,l):
        for e in l:
            if e in self:
                self.remove(e)
        self.sort()
        MySet.eliminate_duplicate(self)        
        return MySet(self)
    
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3,10])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    u=s|t
    print(u)
    u=s&t
    print(u)
    s = MySet([1, 2, 3]) 
    t = MySet([3, 4, 5]) 
    u = s - t 
    print(u) 
```

기존 MySet클래스에서 합집합을 반환하는 메소드를 구현하려면 __or__ 메소드를 추가해 임의의 리스트를 추가한 뒤, 그 리스트에 self 와 l를 append를 이용해 더해주고 그 값을 정렬 한뒤 MySet.eliminate_duplicate을 이용해 중복을 제거해준다. return값은 class인 MySet(lista)으로 해준다.  
교집합을 반환하는 메소드인 __and__ 메소드를 추가해 임의의 리스트를 추가한 뒤, self의 값을 하나씩 불러 그 값이 리스트 l 에 있을 경우 그 값들만 lista에 더해준다. 그값을 정렬한뒤  MySet.eliminate_duplicate을 이용해 중복을 제거해준다. return값은 class인 MySet(lista)으로 해준다.   
차집합을 반환하는 메소드인 __sub__ 메소드를 추가해 뒤에 있는 리스트 l의 원소가 리스트 self에 존재한다면 그 값을 remove해준다.
그값을 정렬한뒤 MySet.eliminate_duplicate을 이용해 중복을 제거해준다. return값은 class인 MySet(self)으로 해준다.
