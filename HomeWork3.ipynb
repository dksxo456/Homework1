그렇기에 [1, 100, 4, 5]값이 된다. 그다임 또 L[1:3] = [(100)]을 하면 100,4가 100으로 치환된다.고로 [1, 100, 5]이 된다.

L[1:3] = 100을 할때 1번째 원소부터 2번째 원소까지 치환하는 것은 문제가 없다.하지만 치환되는 값이 리스트인 []형태가 아니라 그냥 원소만 있으므로 오류가 발생한다. 만약 100대신 100, 을 받으면 튜플형태로 인식하여 문제가 없다.

또, L[1:3] = (100) 또한 치환하는 것은 문제가 없지만 치환되는 값이(100) ()은 튜플의 타입형태 이지만 원소가 1개면 그 원소의 타입을 반환한다.

```python
#문제5
#다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 
#각각 제시하시오.
#1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
def addall(list):
    s=0
    for i in list:
        s+=i
    print(s)
    
addall([1,2,3,4,5,6])
def addall2(list):  
    return sum([x for x in list])
    
    
print(addall2([1,2,3,4,5,6]))
def addallodd(list):
    s=0
    for i in range(len(list)):
        if list[i]%2==1:
            s+=list[i]
    print(s)
          
addallodd([1,2,3,4,5,6,7,8])    
          
def addallodd2(list):
    return sum([list[x] for x in range(len(list)) if list[x]%2==1])
   
print(addallodd2([1,2,3,4,5,6,7,8]))   
```

    addall함수는 리스트에 값들을 모두 더하는 함수 이다. 
    먼저 for~ in리터널 방식을 이용하면 리스트를 반복하면서 하나의 변수s=0을 초기화 하고 그값에 정수를 더해간다.
    리스트 내포 방식을 이용하면 똑같이 리스트를 반복하면서 sum내장함수를 이용해 값들을 합쳐준다.
    addallodd함수는 기존 있던 함수들에 if를 이용하여 홀수인 값들만 더해준다.
    리스트내포방식도 뒤에 if식을 넣어주면 된다. 값이 홀수인 것만 더해줌.
